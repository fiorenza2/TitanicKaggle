1. People with no Parch/SibSp are less likely to survive
2. People with a large number of Parch/SibSp are less likely to survive

Since both of these variables give us an indiciation of family size, let's add the together and see how a new variable 'family size' has an effect on survival:

```python
test = ti_trn.copy()
test['FamSize'] = test['Parch'] + test['SibSp']
plt.plot(test['Survived'].groupby(test['FamSize']).mean())
plt.show()
plt.gcf().clear()
```

```python
test['Survived'].groupby(test['FamSize']).count()
```

This is nice. We can more confidently describe the variables at the tail end (ie: FamSize is 4 or more) as we have more data. Looking at the data, we can say the following:

1. Family sizes of 0 are relatively less likely to live
2. Family sizes of between 1 and 3 inclusive are more likely to live
3. Family sizes of 4 and above are relatively less likely to live.

So lets create a function which returns a variable [N,M,B] for 'None', 'Medium', 'Big' based on the family sizes mentioned above. We'll also remove the Parch/SibSp variables to remove any noise from these:

```python
def getFamSizeNMB(data):
    size = data['SibSp'] + data['Parch']
    data['FamSizeNMB'] = np.where(size == 0, 'N', np.where(size > 3, 'B', 'M'))
    data = data.drop(['SibSp', 'Parch'], axis = 1)
    return(data)
```

```python
test = ti_trn.copy()
test = getFamSizeNMB(test)
test['Survived'].groupby(test['FamSizeNMB']).mean()
```

There's a couple of routes we can go down. We can look at surnames and try to group together families, but that would perhaps create too many variables. We could also look at the titles that exist, which should have much fewer differences. Let us check the significance of this variable:

```python
# takes a list of lists and makes it into 1D
def flattento1D(array):
    flatten = np.array([item for sublist in array for item in sublist])
    return(flatten)
```

Overall, 'Title' gave us some interesting information, so let's use it as a feature, and write a script to add this onto a dataframe, encoding all less common names as 'Other'. Note that Mme. and Mlle. can be mapped to English equivalents(Mrs. and Miss. respectively), so let's take care of that too.

```python
def addTitle(data):
    TitleList = flattento1D(data['Name'].str.findall('[A-z]{2,}\.'))
    TitleList = ['Mrs.' if Title == 'Mme.' else 'Miss.' if (Title == 'Mlle.' or Title == 'Ms.') else Title for Title in TitleList]
    common = ['Mr.', 'Master.', 'Mrs.', 'Miss.']
    TitleList = [Title if Title in common else 'Other' for Title in TitleList]
    data['Title'] = TitleList
    return(data)
```

```python
sns.barplot(x = 'Title', y = 'Survived', data = addTitle(ti_trn))
plt.show()
plt.gcf().clear()
```

```python
def processName(data):
    data = addTitle(data)
    data = data.drop(['Name'], axis = 1)
    return(data)
```

Given the results above, let's make functions which allow us to extract Cabin letter, if it had a cabin field, and whether that number is even:

```python
def CabinInfo(data):
    ExtractLetter = data['Cabin'].str.findall('^[A-Z]')
    ExtractNumber = data['Cabin'].str.findall('([0-9]{1,3})$')
    CabinLetter = [SubList[0] if isinstance(SubList, list) else 'NA' for SubList in ExtractLetter]
    CabinNumber = [SubList if isinstance(SubList, list) else 'NA' for SubList in ExtractNumber]
    CabinNumber = np.array([SubList[0] if (len(SubList) != 0 and isinstance(SubList, list)) else 'NA' for SubList in CabinNumber])
    CabinNoEven = [CheckEven(x) for x in CabinNumber]
    data['CabLetter'] = CabinLetter
    data['CabEven'] = CabinNoEven
    data = data.drop('Cabin', axis = 1)
    return(data)

def CheckEven(x):
    if x == 'NA':
        return x
    elif np.int(x) % 2 == 0:
        return 'Yes'
    else:
        return 'No'
```

### Imputing Age

Having explored the data briefly, we now need to plug the gaps. There are several approaches we can take. In order of relative ease:

1. Drop the rows with missing data
2. Replace with 0s
3. Replace with a statistically derived value for that column (average, median)
4. Same as above, but segmented by different categories (average ages for each Pclass for example)
5. Train a model to calculate the missing values

Note that 2 through 5 are imputation methods.

Method 1 can be the best option, but given how much valuable data we'd be losing, as well as the valuable 'missing age' flag, it's not worth considering here. Of course, we could prove this with a validation test, but I don't think this is needed to prove the point.

Method 2 can give quick and good results, especially if the variable we are considering has no major effect on the variable we are looking to predict. However, in this case age has a demonstrable link to survival rate, so this simply replacing with 0 won't do here. Now we move onto 3 and 4, which are similar, but subtly different.

Method 3 involves calculating a statistic which aggregates the properties of that column, such as average/median/mode. This is an improvement, but we need to remember that the distribution of age is conditional on many different things, such as what class that passenger was in etc.

Method 4 addresses this by setting different aggregate statistics depending on segment. We can therefore first calcluate, for this example, the median ages for female passengers with Pclass 3 and a valid age, and then use that as the replacement age for the gaps where passengers are female and class 3. This allows for better estimates, and due to our aggregation, should not overfit our data too much as long as our segments are wide enough.

Method 5 is the most advanced, and can give the best results. It is highly tempting here, as we have many different descriptors which can be used to form predictions with. However, having done tests with this in a previous attempt, this requires a lot of effort, and the gains were marginal at best (like a 0.2 RMSE gain on average).

Bearing this in mind, let's explore method 4, and see what segments we should condition on.

```python
sns.violinplot(x = 'Pclass', y = 'Age', hue = 'Sex', data = ti_trn)
plt.show()
plt.gcf().clear()
```

## The Model

We'll be using a random forest classifier algorithm for this predictor. This is because of a couple of reasons:

1. Tree-based algorithms work very well with mixed-type data (ie: continuous, integer, categorical)
2. Historically, random forests do well in this competition

The 2nd is a bit meta-referencial, but validates the 1st point. We've already written functions to process the data, so let's get to it. For ease of use I'll be writing a function which does all the processing in one go:

NB: Since we are using a random forest in Python (as opposed to R), we need to convert categorical variables into factors using one-hot encoding. There is built in functionality in pandas for us to do this called get_dummies.

```python
# Processes the raw data input
def processRaw(data):
    data = getFamSizeNMB(data)
    data = processName(data)
    data = CabinInfo(data)
    data = missingEmbark(data)
    data = TickProcess(data)
    data = FillAges(data)
    dummynames = ['Sex', 'Pclass', 'Embarked', 'Title', 'FamSizeNMB', 'CabLetter', 'CabEven']
    data = pd.get_dummies(data, columns = dummynames)
    return(data)
```

With random forests, we can see which features resulted in the most information gain (as these will appear the most in the trees). We can view these on a graph so we can figure out what is really contributing to our model's efficacy. Let us write a wrapper function to help do this:

```python
def plotImp(rndfrst, train):
    importances = rndfrst.feature_importances_
    std = np.std([tree.feature_importances_ for tree in rndfrst.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1]

    plt.figure()
    plt.title("Feature importances")
    plt.bar(range(train.drop('Survived', axis = 1).shape[1]), importances[indices],
           color="r", yerr=std[indices], align="center")
    plt.xticks(range(train.drop('Survived', axis = 1).shape[1]), train.drop('Survived', axis = 1).columns, rotation = 90)
    plt.xlim([-1, train.drop('Survived', axis = 1).shape[1]])
    plt.show()
    plt.gcf().clear()
```

```python
print('OOB Score is ' + str(rfc.oob_score_))
plotImp(rfc, train)
```

How disappointing, as the cabin features seem to be contributing basically nothing! Having said this, our derived pararmeters seem to be working very well.

### Name Length

First time round, we neglected name length. After all, aren't names more or less random? However, people of a certain provedance have longer names (think double barrelling).

Let's figure this out:

```python
def nameLength(data):
    nameLength = [len(Name) for Name in data['Name']]
    data['NameLength'] = nameLength
    return(data)
```

```python
test = ti_trn.copy()
sns.boxplot(x = 'Survived', y = 'NameLength', data = nameLength(test))
plt.show()
plt.gcf().clear()
```

Cool, so it's definitely an indicator here. We could leave it as a continuous variable, or we could encode it. Let's see the distribution:

```python
sns.distplot(nameLength(test)['NameLength'])
plt.show()
plt.gcf().clear()
```

Let's split this up into the various peaks we have. What I mean to say is that let's assume the distrubtion is mixed gaussian model.

```python
def splitnameLength(data):
    TLenCat = ['L' if Len > 41 else 'M' if 25<Len<41 else 'S' for Len in data['NameLength']]
    data['TLenCat'] = TLenCat
    return(data)
```

```python
sns.barplot(x = 'Sex', y = 'Survived', hue = 'TLenCat', data = splitnameLength(nameLength(test)))
plt.show()
sns.barplot(x = 'Pclass', y = 'Survived', hue = 'TLenCat', data = splitnameLength(nameLength(test)))
plt.show()
plt.gcf().clear()
```

This is very interesting. I initially discounted name length as I thought it wouldn't offer anything over Pclass or Sex, but this is clearly not the case. Even controlling for this, the length of someone's name matters!

I think for now we will leave in both the length and category, and see which gets more importance, and modify the processName function to include this:

```python
def nameLenProcess(data):
    data = splitnameLength(nameLength(data))
    return(data)

def processName(data):
    data = addTitle(data)
    data = nameLenProcess(data)
    data = data.drop(['Name'], axis = 1)
    return(data)
```

### Using all the Age Data

As aforementioned, we've only used the age data from the training set, but there is valid data to use in the test set too. Since these are all 'generated' from the same underlying distribution, we can get a better idea of this distribution by using more data. Therefore, we need to create a new 'big' data set of both data, and then calculate the same pivot table to lookup against:

```python
test1 = ti_trn.copy()
test2 = ti_test.copy()

combo = pd.concat([test1.drop('Survived', axis = 1), test2])
combo.info()
```

```python
copy_combo = combo.copy()
processName(copy_combo).pivot_table(values='Age', index=['Title'], columns=['Pclass', 'Sex'], aggfunc=np.median)
```

```python
processName(test1).pivot_table(values='Age', index=['Title'], columns=['Pclass', 'Sex'], aggfunc=np.median)
```

Comparing the median pivots, we can see that there's a fair amount of change, which should hopefully translate into better predictions. With this in mind, let's change the age imputing function:

```python
def combine(train,test):
    return(pd.concat([train.drop('Survived', axis = 1), test]))

def FillAges(data, train, test):
    if 'Title' not in data: # add titles if you couldn't find any, as we need this to segment
        data = addTitle(data)
    def returnAge(row): # create a lookup function given a row of information, included here for scoping purposes
        return(lookupAge[row['Pclass']][row['Sex']][row['Title']])
    combined = combine(train,test)
    combined = processName(combined)
    lookupAge = combined.pivot_table(values='Age', index=['Title'], columns=['Pclass', 'Sex'], aggfunc=np.median) # lookup table
    FilledAges = [Row['Age'] if ~np.isnan(Row['Age']) else returnAge(Row) for Index,Row in data.iterrows()]
    data['Age'] = FilledAges
    return(data)
```

### Mother

When invetigating the box plots surrounding age and survival a while back, we hypothesised that older women survived more than younger women due to the fact that they would most likely have children. Let's explicitely encode this feature and see if we're right. We will define a mother as someone who is:

1. Has Parch greater than 0
2. Is a Mrs.

```python
def isMother(data):
    isMother = [True if (Row['Title'] == 'Mrs.' and Row['Parch'] > 0) else False for Index, Row in data.iterrows()]
    data['Mother'] = isMother
    return(data)
```

Let's check if this gives us something interesting:

### Fare == 0

We can see that some of the data for Fare is down as 0:

```python
ti_trn = pd.read_csv('train.csv')
ti_test = pd.read_csv('test.csv')
```

```python
ti_trn.loc[ti_trn['Fare'] == 0]
```

```python
ti_test.loc[ti_test['Fare'] == 0]
```

So why is this, and how should we fill them? Performing a quick Google, most of these people have a valid reason, such as being crew members who'd been allocated tickets for administrative purposes, or senior shipbuilders sent to oversee the maiden voyage. This is most telling when looking at the SibSp and Parch values, which are all 0. With this in mind, it is appropriate to calculate an imputed Fare value instead. It makes most sense to split by class and look at single male travellers:

```python
sns.boxplot(x = 'Pclass', y = 'Fare', data = ti_trn.loc[(ti_trn['SibSp'] == 0) * (ti_trn['Parch'] == 0) * (ti_trn['Sex'] == 'male')], showfliers = False)
plt.show()
plt.gcf().clear()
```

This is good, we have pretty tight banding of fare prices. Let's shove in the medians where appropriate:

```python
def Calc0Fare(data):
    FareFind = ti_trn.loc[(ti_trn['SibSp'] == 0) & (ti_trn['Parch'] == 0) & (ti_trn['Sex'] == 'male')].copy()
    FareFind = FareFind.pivot_table(values='Fare', columns=['Pclass'], aggfunc=np.median)
    data['Fare'] = [FareFind[Row['Pclass']] if Row['Fare'] == 0 else Row['Fare'] for Index, Row in data.iterrows()]
    return(data)
```

## The Model - Take 2

Cool, so we introduced a few new variables into the mix. Let's redo our models and see if there's any improvement. Note that there was a lot of work done which I didn't document in this notebook. I ended up replacing the Ticket data and didn't include the data for Mother, as it was suppressing the 'Female' variable, causing overfit.

```python
def TickLetLen(data):
    data['TickLet'] = [Ticket[0] for Ticket in data['Ticket']]
    data['TickLet'] = ['Low_Other' if Tick in ['7','W','4','L','6','5','8'] else 
                       'High_Other' if Tick in ['F', '9'] else Tick for Tick in data['TickLet']]
    data['TickLen'] = [len(Ticket) for Ticket in data['Ticket']]
    data = data.drop('Ticket', axis = 1)
    return(data)

def ProcessRawT2(data, train, test):
    data = data.drop('PassengerId', axis = 1)
    data = processName(data)
    data = Calc0Fare(data)
    #data = isMother(data)
    data = getFamSizeNMB(data)
    data = CabinInfo(data)
    data = missingEmbark(data)
    data = TickLetLen(data)
    data = NoAge(data)
    data = FillAges(data, train, test)
    dummynames = ['Sex', 'Pclass', 'Embarked', 'Title', 'FamSizeNMB', 'CabLetter', 'CabEven', 'TLenCat', 'TickLet']
    data = pd.get_dummies(data, columns = dummynames)
    if 'CabLetter_T' in data.columns:
        data = data.drop('CabLetter_T', axis = 1)
    if sum(data['Fare'].isnull()) > 0:
        data['Fare'].loc[data['Fare'].isnull()] = ti_trn['Fare'].median()
    return(data)
```

```python
ti_trn = pd.read_csv('train.csv')
ti_test = pd.read_csv('test.csv')
```

```python
train = ti_trn.copy()
train = ProcessRawT2(train, ti_trn, ti_test)
train.columns
```

```python
test = ti_test.copy()
test = ProcessRawT2(test, ti_trn, ti_test)
```

```python
rfc = RandomForestClassifier(n_estimators= 700, max_features = 4, max_depth = 8, criterion = 'gini', n_jobs = -1, oob_score = True)
cvs = cross_val_score(rfc, train.drop('Survived', axis = 1), train['Survived'], cv = 10)
print('10 fold cross-vaildation error is ' + str(cvs.mean()))
```
